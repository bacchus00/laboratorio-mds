### 2.7 Segmentación Etaria

Agregamos la nueva columnas 'segmentacion_etaria' la cual simplifica la edad de los clientes a una variable categorica de 4 niveles.

```python
def segmentacion_etaria_mapping(x):
  if x < 18: 
    return 'Joven'
  if x in range(18, 27):
    return 'Adulto-Joven'
  if x in range(27, 60):
    return 'Adulto'
  if x >= 60: 
    return 'Persona Mayor'

df['segmentacion_etaria'] = df['customer_age'].apply(segmentacion_etaria_mapping)
df['segmentacion_etaria'].value_counts()
```

## 4. Preprocesamiento

### 4.1 Declarar `ColumnTransformer`

A continuacion definimos los conjuntos de tipos de datos. Dentro de las variables categoricas hay varias variables binarias  que ya vienen codificadas numericamente, la variable 'source' igual es binaria por lo que podria ser util hacer un reemplazo a 0 y 1. Para las demas variables categoricas no hay un orden que se pueda intuir entre sus niveles, por lo que son variables categoricas solo nominales, no ordinales. ES por lo anterior que en estas variables se utilizara 'OneHotEncoding'

```python
df['source'] = df['source'].replace(to_replace={'INTERNET':0, 'TELEAPP':1}).astype(int)

binary_vars = [
    'email_is_free',
    'phone_home_valid',
    'phone_mobile_valid',
    'has_other_cards',
    'foreign_request',
    'keep_alive_session',
    'source'
]

nominal_vars = list(df.select_dtypes(include='object').columns)
numerical_vars = [x for x in list(df.columns) if x not in (binary_vars+nominal_vars+['id', 'fraud_bool'])]
```

```python
print(binary_vars)
print(nominal_vars)
print(numerical_vars)
```

Luego, haciendo uso de ColumnTransformer definimos las transformaciones a realizar sobre las varibales nominales y dejamos pasar el resto de las variables intactas con 'passthrough'.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

t = [("categoricalOHE", OneHotEncoder(sparse_output=False), nominal_vars)]
transformers = ColumnTransformer(transformers=t, remainder='passthrough')
transformers.set_output(transform='pandas')
```

### 4.2 Transformar datos

Ahora se aplica las transformaciones a realizar sobre el dataframe.

```python
transformed_df = transformers.fit_transform(df)
transformed_df.columns
```

### 5.1 Muestrear dataframe

Enseguida se selecciona una muestra aleatoria de 25000 filas y se separa las variables explicativas (X) de la variable target (y)

```python
sample_25k = transformed_df.sample(25000).reset_index(drop=True)

X = sample_25k.loc[:, sample_25k.columns!='remainder__fraud_bool']
y = sample_25k['remainder__fraud_bool']
```

### 5.2 Proyectar y agregar proyecciones al dataframe de muestreo

Recuperamos los indices de las distintas categorias de las segmentacion etaria y extraemos sus variables correpondientes.

```python
i_j = list(X[X['categoricalOHE__segmentacion_etaria_Joven']==1].index)
i_aj = list(X[X['categoricalOHE__segmentacion_etaria_Adulto-Joven']==1].index)
i_a = list(X[X['categoricalOHE__segmentacion_etaria_Adulto']==1].index)
i_pm = list(X[X['categoricalOHE__segmentacion_etaria_Persona Mayor']==1].index)

X_j = X.loc[i_j]
X_aj = X.loc[i_aj]
X_a = X.loc[i_a]
X_pm = X.loc[i_pm]
```
